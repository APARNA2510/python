```python
from dputils import scrape as sc
```

```python
url = 'https://www.amazon.in/s?k=epilator&crid=2DFZM30FLK172&sprefix=epilator%2Caps%2C220&ref=nb_sb_noss_1'
soup = sc.get_webpage_data(url)
```

```python
soup.find('title')
```

```python
target = {
    'tag':'div',
    'attrs':{
        'class':'s-main-slot s-result-list s-search-results sg-row'
    }
}
```

```python
items = {
    'tag':'div',
    'attrs':{
        'class':'s-result-item'
    }
}

title = {
    'tag':'h2',
    'attrs':{
        'class':'a-size-mini a-spacing-none a-color-base s-line-clamp-3'
    }
}

review_count = {
    'tag':'div',
    'attrs':{
        'class':'a-row a-size-small'
    }
}

price = {
    'tag':'span',
    'attrs':{
        'class':'a-price'
    }
}

delivery = {
    'tag':'div',
    'attrs':{
        'class':'a-row a-size-base a-color-secondary s-align-children-center'
    }
}
```

```python
output = sc.extract_many(soup, target=target, items=items, title=title, review_count=review_count, price=price, delivery=delivery)
```
