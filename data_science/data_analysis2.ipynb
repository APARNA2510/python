# data manipulation

```python
# renaming columns
df.rename({'details':'configuration','title':'product_name'},axis = 1, inplace=True)
```

```python
df
```

```python
# adding column
df['website'] = 'flipkart'
```

```python
df
```

```python
# functions to extract rating and reviews
import numpy as np
def extract_rating(value):
    if value:
        value_0 = value.split(' ')[0]
        value_0 = value_0.replace(',','')
        return int(value_0)
    else:
        return np.nan

def extract_reviews(value):
    try:
        return int(value.split()[-2].replace(',',''))
    except:
        return 0                    

# adding columns with function
df['rating'] = df['rr'].apply(extract_rating)    
df['reviews'] = df['rr'].apply(extract_reviews)        
```

```python
df
```

```python
# dropping columns
df.drop(columns=['rr'], inplace=True)
```

```python
df
```

```python
# create a column brand and store the brand name in it taken from th first column
df['brand'] = df['product_name'].apply(lambda x: x.split()[0])
```

```python
df
```

```python
# fixing the column value for example price
df['price'] = df['price'].apply(lambda x: x.replace(',','').replace('₹',''))
```

```python
df['link'] = df['link'].apply(lambda x: 'https://www.flipkart.com'+x)
```

```python
df
```

```python
df.info()
```

```python
df['price'] = df['price'].astype(int)
```

```python
df
```

```python
# sorting data by column
df.sort_values(by='price', ascending=False, inplace=True)
```
