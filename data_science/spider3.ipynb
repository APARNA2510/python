```python
from dputils import scrape as sc
```

```python
url = 'https://www.amazon.in/s?k=thermometer&crid=X14RKNGBGYUC&sprefix=thermo%2Caps%2C205&ref=nb_sb_ss_ts-doa-p_1_6'
soup = sc.get_webpage_data(url)
```

```python
soup.find('title')
```

```python
target = {
    'tag':'div',
    'attrs':{
        'class':'s-main-slot s-result-list s-search-results sg-row'
    }
}
```

```python
items = {
    'tag':'div',
    'attrs':{
        'class':'s-result-item'
    }
}

title = {
    'tag':'h2',
    'attrs':{
        'class':'a-size-mini a-spacing-none a-color-base s-line-clamp-3'
    }
}

reviews = {
    'tag':'div',
    'attrs':{
        'class':'a-row a-size-small'
    }
}

price = {
    'tag':'div',
    'attrs':{
        'class':'a-row a-size-base a-color-base'
    }
}
```

```python
output = sc.extract_many(soup, target=target, items=items, title=title, reviews=reviews, price=price)
```
